```python
# Import pandas for data handling
import pandas as pd

# Import KMeans from scikit-learn
from sklearn.cluster import KMeans

# Import matplotlib for plotting our data
import matplotlib.pyplot as plt

# Import a function to create synthetic data
from sklearn.datasets import make_blobs
```

```python
# Create synthetic data with 4 distinct clusters
X, y = make_blobs(n_samples=200, centers=4, n_features=2,
                  cluster_std=1.2, random_state=42)

# Create a DataFrame for easier handling (optional, but good practice)
df = pd.DataFrame(X, columns=['AnnualIncome', 'SpendingScore'])

# Visualize the unlabeled data
plt.figure(figsize=(8, 6))
plt.scatter(df['AnnualIncome'], df['SpendingScore'], s=50)
plt.title('Customer Data (Unlabeled)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```

```python
# Create an instance of the K-Means model, asking it to find 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)

# Train the model on our data X.
# Note: There is no y! This is unsupervised learning.
kmeans.fit(X)
```

```python
# Get the cluster labels for each data point
labels = kmeans.labels_

# Get the coordinates of the cluster centers
centers = kmeans.cluster_centers_

# Visualize the results
plt.figure(figsize=(8, 6))
# Plot the data points, coloring them by their assigned cluster label
plt.scatter(df['AnnualIncome'], df['SpendingScore'], c=labels, s=50, cmap='viridis')
# Plot the cluster centers as large red 'X's
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Cluster Centers')

plt.title('Customer Segments Found by K-Means')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid(True)
plt.show()

```
